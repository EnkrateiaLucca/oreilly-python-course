## Part 1 · Data Types & Operations

Every value in Python has a **type**. The four you'll use constantly:

| Type | Example | Used for |
|------|---------|----------|
| `int` | `42` | whole numbers (counts, ages, quantities) |
| `float` | `19.99` | decimals (prices, measurements, averages) |
| `str` | `"Ada"` | text (names, emails, messages) |
| `bool` | `True` / `False` | yes/no answers (is it paid? is it valid?) |

Use `type()` to ask Python what something is:

````python
# A tiny customer order, described with the four core types
quantity = 3            # int   — how many
price     = 19.99       # float — how much each
product   = "Notebook"  # str   — the name
in_stock  = True        # bool  — can we sell it?

print(type(quantity))   # <class 'int'>
print(type(price))      # <class 'float'>
print(type(product))    # <class 'str'>
print(type(in_stock))   # <class 'bool'>
````

### Numbers: arithmetic

Python is a calculator that never gets tired. A quick **bill splitter**:

````python
bill      = 84.50   # dinner total
people    = 3       # splitting between friends
tip_rate  = 0.20    # 20% tip

total      = bill + (bill * tip_rate)   # add the tip
per_person = total / people             # split it

print(total)        # 101.4
print(per_person)   # 33.8
````

A few operators that surprise beginners:

| Operator | Name | `17` with `5` | Handy for |
|----------|------|---------------|-----------|
| `/` | divide | `3.4` | normal division (always a float) |
| `//` | floor divide | `3` | "how many whole groups?" |
| `%` | modulo | `2` | "what's left over?" / "is it even?" |
| `**` | power | `17**2 = 289` | squares, growth |

````python
# Real use: 100 eggs into cartons of 12
eggs = 100
print(eggs // 12, "full cartons")   # 8 full cartons
print(eggs %  12, "eggs left over") # 4 eggs left over

# Is a number even?  (even numbers have remainder 0 when divided by 2)
print(100 % 2 == 0)   # True  -> 100 is even
````

### Strings: working with text

Text is *combined* with `+` and *repeated* with `*`:

````python
first = "Ada"
last  = "Lovelace"

full_name = first + " " + last      # join with a space
print(full_name)                    # Ada Lovelace

print("-" * 20)                     # a quick divider line
print(full_name.upper())            # ADA LOVELACE  (a string "method")
print(len(full_name))               # 12  (number of characters)
````

> ⚠️ **A classic beginner error.** You can't add text to a number directly —
> Python doesn't know if you mean "glue them" or "do math". Run this and read the error;
> learning to *read errors* is a real skill.

````python
"You have " + 100 + " points"   # TypeError on purpose — fix it in the next cell
````

````python
# Fix 1: convert the number to a string with str(...)
print("You have " + str(100) + " points")

# Fix 2 (better): an f-string — see below
print(f"You have {100} points")
````

### f-strings: the best way to build text

Put an `f` before the quotes, then drop any variable inside `{curly braces}`. This is how
you'll format almost every message, label, and report in this course.

````python
name   = "Sam"
items  = 3
total  = 47.5

# Build a real order-confirmation line
print(f"Hi {name}, your {items} items come to ${total:.2f}.")
# Hi Sam, your 3 items come to $47.50.

# {total:.2f} means "show 2 decimal places" — perfect for money
````

## Part 2 · Variables

A variable is a **labelled box** that holds a value. You put something in, and refer to it
later by its label.

```
   product            price            in_stock
 ┌──────────┐      ┌─────────┐       ┌────────┐
 │ "Coffee" │      │  4.50   │       │  True  │
 └──────────┘      └─────────┘       └────────┘
```

Good names make code read like English. Prefer `total_price` over `tp`.

````python
# Variables can be updated — a shopping cart total growing as you add items
cart_total = 0          # start empty

cart_total = cart_total + 4.50   # add a coffee
cart_total += 2.25               # += is shorthand for "add to itself"
cart_total += 9.99               # add a sandwich

print(f"Cart total: ${cart_total:.2f}")   # Cart total: $16.74
````

### Getting input from a person

`input()` pauses and waits for the user to type. It always returns a **string**, so wrap numbers in `int()` or `float()` before doing math.

````python
# A mini tip calculator (run it and type a number when prompted)
bill = float(input("What was the bill? $"))   # float() -> turn the text into a number
tip  = bill * 0.18

print(f"Add a $18% tip of ${tip:.2f}")
print(f"Total to pay: ${bill + tip:.2f}")
````

## Part 3 · Functions

A function is a **named, reusable recipe**. Define it once, call it as many times as you like.

```
   inputs              the recipe                output
  (arguments) ──►   def add_tip(bill):   ──►   return value
                        ...
```

Why bother? So you never copy-paste the same logic twice.

````python
# Useful built-in functions you'll reach for constantly
prices = [4.50, 2.25, 9.99, 1.75]

print(len(prices))        # 4      — how many items
print(sum(prices))        # 18.49  — add them all
print(max(prices))        # 9.99   — most expensive
print(min(prices))        # 1.75   — cheapest
print(round(18.49, 1))    # 18.5   — round to 1 decimal
````

### Writing your own function

The tip calculator from before, but reusable — and it handles *any* bill and tip rate:

````python
def add_tip(bill, tip_rate=0.18):
    """Return the bill total including a tip. Tip defaults to 18%."""
    return bill + (bill * tip_rate)

# Call it with different inputs — no copy-paste
print(add_tip(50))          # 59.0   (uses the default 18%)
print(add_tip(50, 0.20))    # 60.0   (override to 20%)
print(add_tip(120, 0.15))   # 138.0
````

### A function that returns labelled results

Return a **dictionary** when a function produces several related answers — like a quick stats report on a list of numbers:

````python
def summarize(numbers):
    """Return basic stats about a list of numbers."""
    return {
        "count":   len(numbers),
        "total":   sum(numbers),
        "average": sum(numbers) / len(numbers),
        "highest": max(numbers),
        "lowest":  min(numbers),
    }

scores = [85, 92, 78, 90, 88]
report = summarize(scores)

print(f"Average score: {report['average']:.1f}")   # Average score: 86.6
print(f"Top score:     {report['highest']}")        # Top score:     92
````

## Part 4 · Lists & Loops

A **list** holds many values in order. A **loop** lets you do something with each one.

```
 index:     0        1        2        3
         ┌───────┬────────┬───────┬────────┐
 todos = │ "Buy" │ "Call" │ "Pay" │ "Read" │
         └───────┴────────┴───────┴────────┘
              todos[0] = "Buy"   todos[-1] = "Read" (last)
```

````python
# A to-do list, indexed from 0
todos = ["Buy groceries", "Call dentist", "Pay rent", "Read 10 pages"]

print(todos[0])     # Buy groceries   (first)
print(todos[-1])    # Read 10 pages   (last — negative counts from the end)
print(len(todos))   # 4               (how many tasks)
````

### Changing a list

Lists grow and shrink. A shopping list in action:

````python
shopping = ["milk", "bread", "eggs"]

shopping.append("cheese")              # add one to the end
shopping.extend(["apples", "rice"])    # add several
shopping.remove("bread")               # take a specific item out

print(shopping)            # ['milk', 'eggs', 'cheese', 'apples', 'rice']
print("milk" in shopping)  # True — handy membership check
````

### `for` loops: do something with each item

This is where lists become powerful. A numbered task tracker:

````python
todos = ["Buy groceries", "Call dentist", "Pay rent"]

for number, task in enumerate(todos, start=1):
    print(f"{number}. [ ] {task}")

# 1. [ ] Buy groceries
# 2. [ ] Call dentist
# 3. [ ] Pay rent
````

````python
# Loops + math: total up a shopping cart
cart = [4.50, 2.25, 9.99, 1.75]

running_total = 0
for price in cart:
    running_total += price        # add each price as we go

print(f"You're buying {len(cart)} items for ${running_total:.2f}")
# You're buying 4 items for $18.49
````

## Part 5 · Dictionaries

A list stores values *by position*. A **dictionary** stores values *by label* (a "key").
Perfect when each piece of data has a name.

```
        key            value
   ┌───────────┬──────────────┐
   │ "Alice"   │ "555-0101"   │
   │ "Bob"     │ "555-0202"   │
   │ "Charlie" │ "555-0303"   │
   └───────────┴──────────────┘
```

````python
# A contact book: look people up by name instead of by number
contacts = {
    "Alice": "555-0101",
    "Bob":   "555-0202",
    "Charlie": "555-0303",
}

print(contacts["Alice"])                  # 555-0101
print(contacts.get("Eve", "Not found"))   # .get() avoids a crash if the key is missing
````

### Looping over a dictionary

`.items()` gives you each key and value together — great for inventories, settings, counts:

````python
inventory = {"apples": 50, "bananas": 30, "oranges": 25}

# Restock and report
inventory["apples"] = 45        # update an existing item
inventory["grapes"] = 40        # add a new one

for fruit, quantity in inventory.items():
    status = "LOW" if quantity < 30 else "ok"
    print(f"{fruit:>8}: {quantity:>3}  ({status})")
````

### A real pattern: counting things

Dictionaries are the natural way to **tally** — e.g. count how often each word appears:

````python
sentence = "the cat sat on the mat the cat purred"

counts = {}
for word in sentence.split():               # .split() -> list of words
    counts[word] = counts.get(word, 0) + 1  # add 1, starting from 0

print(counts)
# {'the': 3, 'cat': 2, 'sat': 1, 'on': 1, 'mat': 1, 'purred': 1}
````

### Nested dictionaries

Values can themselves be dictionaries or lists — that's how real data (API responses, config files) is shaped:

````python
user = {
    "name": "Sarah Wilson",
    "role": "Software Engineer",
    "skills": ["Python", "SQL", "Docker"],
    "settings": {"theme": "dark", "notifications": True},
}

print(user["name"])                       # Sarah Wilson
print(user["skills"][0])                   # Python  (first skill)
print(user["settings"]["theme"])           # dark    (dict inside a dict)
````

## Part 6 · Comparisons & Logic

Comparisons answer yes/no questions and return a **bool** (`True`/`False`). They're the
raw material of every decision your program makes.

| Operator | Means | Example | Result |
|----------|-------|---------|--------|
| `==` | equal to | `age == 18` | |
| `!=` | not equal | `status != "paid"` | |
| `>` `<` | greater / less | `score > 70` | |
| `>=` `<=` | greater/less or equal | `qty >= 1` | |

````python
# Validating a sign-up form
age      = 16
password = "secret123"

print(age >= 18)            # False — old enough?
print(len(password) >= 8)   # True  — long enough?
print(password == "")       # False — was it left blank?
````

### Combining conditions: `and`, `or`, `not`

- **`and`** → *all* must be true
- **`or`** → *at least one* must be true
- **`not`** → flips it

| A | B | `A and B` | `A or B` |
|---|---|-----------|----------|
| True | True | True | True |
| True | False | False | True |
| False | False | False | False |

````python
# Can this user check out?  Cart must be non-empty AND payment must be valid.
cart_items   = 3
payment_ok   = True

can_checkout = cart_items > 0 and payment_ok
print(can_checkout)          # True

# Free shipping if you spend $50+  OR  you're a member
subtotal  = 30
is_member = True
free_shipping = subtotal >= 50 or is_member
print(free_shipping)         # True
````

## Part 7 · Conditionals — making decisions

`if` / `elif` / `else` run different code depending on the answer to a comparison.

```
            ┌─ if condition:      run this …
 decision ──┼─ elif other:        … or this …
            └─ else:              … otherwise this
```

````python
# Low-battery warning
battery = 15

if battery < 20:
    print("🔋 Battery low — plug in your charger!")
else:
    print("🔋 Battery fine.")
````

### `if` / `elif` / `else` chains

Checked top to bottom; the **first** true branch wins. A letter-grade converter:

````python
def letter_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    else:
        return "F"

for s in [95, 82, 71, 60]:
    print(f"{s} -> {letter_grade(s)}")
# 95 -> A,  82 -> B,  71 -> C,  60 -> F
````

## 🏆 Putting it all together: a Smart Downloads Organizer

This little script uses **everything** from this notebook — functions, strings, comparisons,
boolean logic, dictionaries, lists, and `if`/`elif`/`else` — to decide where each file in a
messy Downloads folder should go. This is the *real* shape of the automation you'll build in
Day 2.

````python
def organize_file(filename, size_mb):
    """Decide a destination folder and an action for one file."""
    name = filename.lower()   # so ".JPG" and ".jpg" are treated the same

    # 1) Choose a folder from the file extension
    if name.endswith((".jpg", ".jpeg", ".png", ".gif")):
        folder = "Images"
    elif name.endswith((".mp4", ".mov", ".avi")):
        folder = "Videos"
    elif name.endswith((".pdf", ".docx", ".txt", ".md")):
        folder = "Documents"
    elif name.endswith((".zip", ".tar", ".gz")):
        folder = "Archives"
    else:
        folder = "Other"

    # 2) Yes/no flags from comparisons + string methods
    is_large = size_mb > 100
    is_temp  = name.startswith("~") or name.endswith(".tmp")

    # 3) Combine the flags to choose an action
    if is_temp:
        action = "🗑️  Delete (temporary file)"
    elif is_large:
        action = f"📦 Move to {folder} (⚠️ large: {size_mb} MB)"
    else:
        action = f"✅ Move to {folder}"

    return {"file": filename, "folder": folder, "action": action}


# A messy folder: a list of {name, size} dictionaries
downloads = [
    {"name": "vacation_photo.JPG", "size": 4},
    {"name": "course_slides.pdf",  "size": 12},
    {"name": "movie_trailer.mp4",  "size": 350},
    {"name": "project_backup.zip", "size": 80},
    {"name": "~draft.tmp",         "size": 1},
    {"name": "mystery_file.xyz",   "size": 2},
]

print("📁 Organizing your Downloads folder...\n")

folder_counts = {}   # tally how many files land in each folder
for item in downloads:
    result = organize_file(item["name"], item["size"])
    print(f"{result['file']}\n   {result['action']}\n")
    folder = result["folder"]
    folder_counts[folder] = folder_counts.get(folder, 0) + 1

print("Summary — files per folder:")
for folder, count in folder_counts.items():
    print(f"  {folder}: {count}")
````

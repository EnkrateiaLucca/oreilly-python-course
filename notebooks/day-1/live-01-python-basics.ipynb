## Part 3 · Functions

A function is a **named, reusable recipe**. Define it once, call it as many times as you like.

```
   inputs              the recipe                output
  (arguments) ──►   def add_tip(bill):   ──►   return value
                        ...
```

Why bother? So you never copy-paste the same logic twice.

````python
# FUnctions are reusable blocks of code
def this_is_a_function():
    name = "Lucas"
    print("Whose name?")
    print(name)
````

````python
this_is_a_function()
````

````python
this_is_a_function()
````

````python
this_is_a_function()
````

````python
def process_product_catalog(product_name, product_price, in_stock_info):
    product_catalog_info = f"Product name: {product_name};Price: {product_price};In Stock?{in_stock_info}"
    
    return product_catalog_info

name = "JetPack"
price = 300000
in_stock = True
# I can put them as they are in the right sequence
print(process_product_catalog(name, price, in_stock))
# I can also name the parameters and assign them to the corresponding variables
print(process_product_catalog(product_name=name, product_price=price, in_stock_info=in_stock))
````

## Part 5 · Dictionaries

A list stores values *by position*. A **dictionary** stores values *by label* (a "key").
Perfect when each piece of data has a name.

```
        key            value
   ┌───────────┬──────────────┐
   │ "Alice"   │ "555-0101"   │
   │ "Bob"     │ "555-0202"   │
   │ "Charlie" │ "555-0303"   │
   └───────────┴──────────────┘
```

Imagine you wanted to store a contacts_list with names and the phone numbers.
If you were using lists you would do something like this:

````python
contacts_list = ["John", "TQ", "William", "Lucas"]
phone_numbers = ["555-1234", "555-3232", "555-1111", "555-723"]

phone_numbers[contacts_list.index("John")]
````

````python
# what is John's phone number?
# This would be the bad way, using lists
def get_contact_phone_number(name):
    contacts_list = ["John", "TQ", "William", "Lucas"]
    phone_numbers = ["555-1234", "555-3232", "555-1111", "555-723"]
    phone_num = phone_numbers[contacts_list.index(name)]
    return phone_num
get_contact_phone_number("Lucas")
````

````python
contacts_dictionary = {"John":"555-1234",
                        "TQ":"555-3232",
                        "William":"555-1111",
                        "Lucas":"555-723"}

contacts_dictionary.get("Lucas")
````

## Part 6 · Comparisons & Logic

Comparisons answer yes/no questions and return a **bool** (`True`/`False`). They're the
raw material of every decision your program makes.

| Operator | Means | Example | Result |
|----------|-------|---------|--------|
| `==` | equal to | `age == 18` | |
| `!=` | not equal | `status != "paid"` | |
| `>` `<` | greater / less | `score > 70` | |
| `>=` `<=` | greater/less or equal | `qty >= 1` | |

### Combining conditions: `and`, `or`, `not`

- **`and`** → *all* must be true
- **`or`** → *at least one* must be true
- **`not`** → flips it

| A | B | `A and B` | `A or B` |
|---|---|-----------|----------|
| True | True | True | True |
| True | False | False | True |
| False | False | False | False |

## Part 7 · Conditionals — making decisions

`if` / `elif` / `else` run different code depending on the answer to a comparison.

```
            ┌─ if condition:      run this …
 decision ──┼─ elif other:        … or this …
            └─ else:              … otherwise this
```

````python
# Imagine folks I was building a prioritization system for purchases
threshold_amount = 1000
price_of_the_thing = input("What is the price of the product you are considering?")

if int(price_of_the_thing)<threshold_amount:
    print("Buy the thing!")
else:
    print("Don't buy the thing!")
````

### `if` / `elif` / `else` chains

Checked top to bottom; the **first** true branch wins. A letter-grade converter:

````python
def letter_grade_converter(student_grade):
    letter_grade = ""
    if student_grade <= 10 and student_grade>=8:
        letter_grade = "A"
    elif student_grade<8 and student_grade>=6:
        letter_grade = "B"
    else:
        letter_grade = "C"
    
    return letter_grade

letter_grade_converter(10)
````

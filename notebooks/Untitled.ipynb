## Time Travel Calculator

- ask the user for their current age (integer/whole number)
- ash how many years into the future they want to travel (integer)
- calculate and print their future age using f-strings.

```python
user_age = input("What is your age?")

type(user_age)
```

```python
user_age = input("What is your age?")

user_age_integer = int(user_age) # here folks I am converting the string into an integer number

future_years = input("How many years into the future you would like to travel? (output an integer: e.g 10)")

future_years_integer = int(future_years) # converts string to integer

future_age = user_age_integer + future_years_integer

print(f"Your future age will be: {future_age} years old.")
```

# Functions

Its a way for us to bundle code together to perform a task like the time travel calculator. 

```python
def time_travel_calculator():
    # (btw this is a comment!)
    # all the steps for the function
    user_age = int(input("what is your age?"))
    future_travel = int(input("How far into the future?"))
    future_age = user_age + future_travel
    print(f"Your future age will be: {future_age}")
    return future_age
```

```python
time_travel_calculator()
```

## Python Logical Operators

### `and` Operator

| A       | B       | A and B |
|---------|---------|---------|
| True    | True    | True    |
| True    | False   | False   |
| False   | True    | False   |
| False   | False   | False   |

### `or` Operator

| A       | B       | A or B  |
|---------|---------|---------|
| True    | True    | True    |
| True    | False   | True    |
| False   | True    | True    |
| False   | False   | False   |

### `not` Operator

| A       | not A   |
|---------|---------|
| True    | False   |
| False   | True    |

Imagine we want to check the file status on our machines with a custom python function.

```python
def check_file_status(file_size_mb):
    if file_size_mb==0:
        print("Error file empty")
    elif file_size_mb < 1:
        print("Warning: file is super small!")
    elif file_size_mb > 1000:
        print("Warning file is too large!")
    else:
        print("File size is acceptable!")

check_file_status(0.2)
check_file_status(1)
check_file_status(200)
check_file_status(1001)
```

```python
def classify_file_type(file_name: str):
    if file_name.endswith(".txt"):
        print("Text file!")
    elif file_name.endswith(".csv"):
        print("CSV File!")
    elif file_name.endswith(".zip"):
        print("Zip file!")
    else:
        print("Not a known file")


classify_file_type("file.txt")
classify_file_type("file.csv")
classify_file_type("file.zip")
classify_file_type("file.mdr")
```

## Let's build some sort of automation!

```python
# I want an automation to buy an item if it costs less than a certain amount say, 5$.
def check_price_and_buy(item, price_threshold):
    """
    It takes an item, checks its price, if its below the price threshoold
    it will 'fake' buy it buy printing buy.
    """
    
    shopping_items = {"apple": 2, "banana": 1, "toaster": 10}

    if shopping_items.get(item): # we are checking if the item exists!
        price_item = shopping_items[item]
        if price_item<price_threshold:
            print("Buy it!")
        else:
            print("No buy!")
    return "checking done!"

check_price_and_buy("apple", 5)
check_price_and_buy("banana", 5)
check_price_and_buy("toaster", 5)
check_price_and_buy("oansdojgnas;dgnj", 5)
```
